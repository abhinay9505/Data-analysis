# What is the average Delivery Time taken by delivery boys to deliver the Pizza?


```python
df['delivery_time'] = (df['order_delivered_at'] - df['order_placed_at']).dt.total_seconds() / 60 

average_delivery_time = df['delivery_time'].mean()

print("The average delivery time taken by delivery boys to deliver pizza is:", average_delivery_time, "minutes")
```

# On which week day have you observed maximum orders?


```python
df['weekday'] = df['order_placed_at'].dt.weekday
weekday_counts = df['weekday'].value_counts()# count of orders for each weekday

max_orders_weekday = weekday_counts.idxmax()# weekday maximum orders
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"] #Convert weekday name
max_orders_weekday_name = weekday_names[max_orders_weekday]

print("The weekday with the maximum orders is:", max_orders_weekday_name)
```

# How many pizza's got delivered after 30 minutes and what is the total amount for such orders?
*
This is important as Diminos promise 30 minutes delivery. For any pizza delivered after 30 minutes, customer will get a full refund.

```python
df['actual_delivery_time'] = df['order_delivered_at'] - df['order_placed_at']

late_deliveries = df[df['actual_delivery_time'] > pd.Timedelta(minutes=30)]

amount_of_deliveries = late_deliveries['order_amount'].sum()#total amount for late deliveries

num_late_deliveries = len(late_deliveries)# Get the count of late deliveries

print("Number of pizzas delivered after 30 minutes:", num_late_deliveries)
print("Total amount for late deliveries:", amount_of_deliveries)
```

# How many pizza's were ordered between 1st March and 5th March 2023?
